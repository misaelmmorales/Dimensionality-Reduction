```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = np.load('facies2999.npy')
f.shape
```

```python
fv = f.reshape(1000,-1)

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=49)
pca.fit(fv)
z = pca.transform(fv)
z.shape
```

```python
fh = z.reshape(1000,7,7)
fh.shape
```

```python
plt.figure(figsize=(16,10))
k = 1
for i in range(10):
    for j in range(10):
        plt.subplot(10,10,k)
        plt.imshow(fh[k], cmap='jet')
        plt.xticks([]); plt.yticks([])
        k += 1
plt.tight_layout()
plt.show()
```
